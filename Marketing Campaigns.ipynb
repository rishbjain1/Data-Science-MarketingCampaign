```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define numerical columns to visualize
columns_to_plot = ["Income", "Age", "Total_Spent", "Total_purchases", "Total_children"]

# Function to remove outliers
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Remove outliers for each column
for col in columns_to_plot:
    df = remove_outliers(df, col)

# Generate Histograms
for col in columns_to_plot:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], bins=30, kde=True)  # Histogram with Kernel Density Estimate (KDE)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Generate Box Plots
for col in columns_to_plot:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])  # Box plot to visualize outliers
    plt.title(f"Box Plot of {col}")
    plt.xlabel(col)
    plt.show()
```

Hypothesis Testing 


a.	Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.

1. Define the Hypotheses:

Null Hypothesis (H₀): Older individuals (Age ≥ 53) do not significantly prefer in-store shopping over online shopping.

Alternative Hypothesis (H₁): Older individuals (Age ≥ 53) significantly prefer in-store shopping over online shopping.

```python
import pandas as pd
from scipy.stats import norm
import numpy as np

# Assuming df1 is your DataFrame and it has columns 'Age', 'NumStorePurchases', and 'NumWebPurchases'
# df1 = pd.read_csv('your_data.csv')

# Define the age threshold
age_threshold = 53

# Filter the DataFrame to include only older individuals
older_individuals = df1[df1['Age'] >= age_threshold]

# Calculate the mean and standard deviation for store and web purchases
mean_store_purchases = older_individuals['NumStorePurchases'].mean()
std_store_purchases = older_individuals['NumStorePurchases'].std()
n_store = len(older_individuals['NumStorePurchases'])

mean_web_purchases = older_individuals['NumWebPurchases'].mean()
std_web_purchases = older_individuals['NumWebPurchases'].std()
n_web = len(older_individuals['NumWebPurchases'])

# Print the results
print(f"Mean number of store purchases: {mean_store_purchases}")
print(f"Standard deviation of store purchases: {std_store_purchases}")
print(f"Number of store purchase observations: {n_store}")

print(f"Mean number of web purchases: {mean_web_purchases}")
print(f"Standard deviation of web purchases: {std_web_purchases}")
print(f"Number of web purchase observations: {n_web}")

# Calculate the z-statistic for the two-sample z-test
pooled_std = np.sqrt((std_store_purchases**2 / n_store) + (std_web_purchases**2 / n_web))
z_stat = (mean_store_purchases - mean_web_purchases) / pooled_std
p_value = 1 - norm.cdf(z_stat)

# Print the z-test results
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Decision-making
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. Older individuals significantly prefer in-store shopping over online shopping.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence that older individuals prefer in-store shopping over online shopping.")
```

b.	Customers with children likely experience time constraints, making online shopping a more convenient option.

Null Hypothesis (H₀): Customers with children do not find online shopping more convenient than in-store shopping.
μ_web = μ_store

Alternative Hypothesis (H₁): Customers with children find online shopping more convenient than in-store shopping.
μ_web > μ_store


```python
import numpy as np
import pandas as pd
import scipy.stats as stats



# Step 1: Create a Column for Customers with Children (Kidhome + Teenhome > 0)
df1['Has_Children'] = (df1['Kidhome'] + df1['Teenhome']) > 0

# Step 2: Filter Only Customers with Children
children_df = df1[df1['Has_Children'] == True]

# Step 3: Compute the Difference Between Online and In-Store Purchases
children_df['Difference'] = children_df['NumWebPurchases'] - children_df['NumStorePurchases']

# Step 4: Compute Sample Statistics
mean_diff = np.mean(children_df['Difference'])  # Mean difference
std_diff = np.std(children_df['Difference'], ddof=1)  # Sample standard deviation of differences
n = len(children_df)  # Sample size

# Known population standard deviation (replace if known)
pop_std = None  # If known, replace with actual population std dev

# Step 5: Compute Z-score (if population std is known)
if pop_std:
    standard_error = pop_std / np.sqrt(n)
    z_stat = mean_diff / standard_error
    p_value = 1 - stats.norm.cdf(z_stat)  # One-tailed test
    print(f"Paired Z-Test: Z-Statistic = {z_stat}, p-value = {p_value}")
else:
    # If population standard deviation is unknown, use paired t-test instead
    t_stat, t_p_value = stats.ttest_rel(children_df['NumWebPurchases'], children_df['NumStorePurchases'])
    p_value = t_p_value / 2  # Convert to one-tailed p-value
    print(f"Paired t-Test (default): t-Statistic = {t_stat}, p-value = {p_value}")

# Step 6: Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Result: Reject H₀. Customers with children significantly prefer online shopping over in-store shopping.")
else:
    print("Result: Fail to reject H₀. No significant preference for online shopping among customers with children.")
```

c.	Sales at physical stores may face the risk of cannibalization by alternative distribution channels.

Define the Hypotheses

Null Hypothesis (H₀): Sales at physical stores are not significantly impacted by alternative distribution channels (online and catalog purchases).
μ_store = μ_web + μ_catalog

Alternative Hypothesis (H₁): Sales at physical stores are significantly cannibalized by alternative distribution channels (online and catalog purchases).
μ_store < μ_web + μ_catalog


```python
import numpy as np
import scipy.stats as stats

# Step 1: Compute Total Alternative Purchases (Web + Catalog)
df1['Alternative_Purchases'] = df1['NumWebPurchases'] + df1['NumCatalogPurchases']

# Step 2: Compute the Difference (Alternative Purchases - Store Purchases)
df1['Difference'] = df1['Alternative_Purchases'] - df1['NumStorePurchases']

# Step 3: Compute Sample Statistics
mean_diff = np.mean(df1['Difference'])  # Mean difference
std_diff = np.std(df1['Difference'], ddof=1)  # Sample standard deviation of differences
n = len(df1)  # Sample size


# Known population standard deviation (replace if known)
pop_std = None  # If known, replace with actual population std dev

# Step 4: Compute Z-score (if population std is known)
if pop_std:
    standard_error = pop_std / np.sqrt(n)
    z_stat = mean_diff / standard_error
    p_value = stats.norm.cdf(z_stat)  # One-tailed test (store purchases < alternative purchases)
    print(f"Paired Z-Test: Z-Statistic = {z_stat}, p-value = {p_value}")
else:
    # If population standard deviation is unknown, use paired t-test instead
    t_stat, t_p_value = stats.ttest_rel(df1['NumStorePurchases'], df1['Alternative_Purchases'])
    p_value = t_p_value / 2  # Convert to one-tailed p-value
    print(f"Paired t-Test (default): t-Statistic = {t_stat}, p-value = {p_value}")

# Step 5: Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Result: Reject H₀. Physical store sales are significantly cannibalized by alternative distribution channels.")
else:
    print("Result: Fail to reject H₀. No significant cannibalization of physical store sales.")
```

d.	Does the United States significantly outperform the rest of the world in total purchase volumes?



Null Hypothesis (H₀): The United States does not significantly outperform the rest of the world in total purchase volume.

​	Alternative Hypothesis (H₁): The United States significantly outperforms the rest of the world in total purchase volume.

```python
import numpy as np
import scipy.stats as stats

# Step 1: Compute Total Purchase Volume for Each Customer
df['TotalPurchase'] = (
    df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + 
    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
)

# Step 2: Separate Customers from the U.S. and the Rest of the World
us_customers = df[df['Country'] == 'United States']
world_customers = df[df['Country'] != 'United States']

# Step 3: Compute Statistics for Total Purchases
mean_us = np.mean(us_customers['TotalPurchase'])
std_us = np.std(us_customers['TotalPurchase'], ddof=1)  # Sample std deviation
n_us = len(us_customers)

mean_world = np.mean(world_customers['TotalPurchase'])
std_world = np.std(world_customers['TotalPurchase'], ddof=1)  # Sample std deviation
n_world = len(world_customers)

# Step 4: Compute the Z-score
if n_us > 0 and n_world > 0:  # Check if denominators are non-zero
    pooled_std = np.sqrt((std_us**2 / n_us) + (std_world**2 / n_world))
    z_score = (mean_us - mean_world) / pooled_std
else:
    z_score = 0  # Set z_score to 0 if denominators are zero

# Step 5: Compute the p-value for a one-tailed test (US > Rest of the World)
p_value = 1 - stats.norm.cdf(z_score)  # One-tailed test

# Print results
print(f"Two-Sample Z-Test:")
print(f"Mean Total Purchase (US) = {mean_us}, Mean Total Purchase (World) = {mean_world}")
print(f"Z-Statistic = {z_score}, p-value = {p_value}")

# Step 6: Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Result: Reject H₀. The United States significantly outperforms the rest of the world in total purchase volumes.")
else:
    print("Result: Fail to reject H₀. No significant difference in total purchase volumes between the U.S. and the rest of the world.")
```
