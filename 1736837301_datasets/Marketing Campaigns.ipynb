```python
import pandas as pd

import matplotlib.pyplot as plt

import datetime as dt

# Import CSV file
df = pd.read_csv('marketing_data.csv')

df
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format= 'mixed', errors='coerce') 
```

```python
df['Dt_Customer']
```

```python
df[' Income '] = df[' Income '].str.replace('$', '')

df = df.rename(columns={' Income ': 'Income'})

```

```python
#df['Income'].fillna(0, inplace=True)
```

```python
df['Income'].value_counts()
```

```python
df['Income'].info()

df[df['Income'].isnull()]
```

```python
df.drop(df[df['Marital_Status'].isin(['YOLO', 'Absurd'])].index, inplace=True)
df['Marital_Status'] = df['Marital_Status'].replace('Alone', 'Single')

```

```python
df['Education'].value_counts()
```

```python
df['Education'] = df['Education'].replace('2n Cycle', 'Second Cycle')
```

```python
df['Education'].value_counts()
```

```python
df['Marital_Status'].value_counts()
```

```python
df['Income'] = df['Income'].str.replace(',', '')
```

```python
df['Income'] = pd.to_numeric(df['Income'], errors='coerce').astype('Int64')
```

```python
df[df['Income'].isnull()]
```

```python
print("Missing Income values before imputation:", df["Income"].isnull().sum())
```

```python
# Convert 'Income' to numeric and then to Int64
df['Income'] = pd.to_numeric(df['Income'], errors='coerce').astype('float')

# Fill missing 'Income' values with the mean 'Income' of the respective 'Education' and 'Marital_Status' groups
df['Income'] = df['Income'].fillna(df.groupby(['Education', 'Marital_Status'])['Income'].transform('mean'))

print(df)
```

```python
print("Missing Income values before imputation:", df["Income"].isnull().sum())
```

```python
df['Income'] = df['Income'].apply(lambda x: pd.NA if pd.isna(x) else int(x))
```
